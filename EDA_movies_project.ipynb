```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/Movies Dataset.csv')
movies.head(2)
```

```python
movies.info()
```

```python
movies.drop(columns=["production_companies"	,"overview" , "tagline" , "credits" , "keywords" , "poster_path" , "backdrop_path" , "recommendations"] ,axis =1 , inplace = True )
```

```python
movies.info()
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isna().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.drop_duplicates(inplace=True)
```

```python
movies["release_date"]=pd.to_datetime(movies['release_date'])
```

```python
movies.info()
```

```python
movies.duplicated().sum()
```

```python
movies["genres"].value_counts()
```

```python
def extract_first_genre(genres_str):

    if pd.isna(genres_str) or genres_str == '':
        return None
    elif '-' in genres_str:
        return genres_str.split('-')[0]
    else:
        return genres_str

movies['genres'] = movies['genres'].apply(extract_first_genre)
```

```python
movies["genres"].value_counts()
```

```python
movies["title"].value_counts()
```

```python
movies.drop_duplicates(subset=['title'], inplace=True)
movies["title"].value_counts()
```

```python
index_to_drop = movies[movies['runtime'] == 0].index
movies.drop(index_to_drop, inplace=True)
```

```python
movies.info()
```

```python
movies.describe()
```

```python
movies.describe(include = 'object')
```

```python
#the average runtime of all movies
average_runtime = np.mean(movies['runtime']).round(0)
print(f"Average runtime: {average_runtime} minutes")
```

```python
#the highest budget for a movie
highest_budget = movies['budget'].max()
highest_budget
```

```python
 #movies were released in each year
movies['release_year'] = movies['release_date'].dt.year
yearly_counts = movies['release_year'].value_counts().sort_values(ascending=False).head(10)
yearly_counts
```

```python
#the total revenue generated by all movies
total_revenue = movies['revenue'].sum()
total_revenue
```

```python
#the median popularity score of movies
median_popularity = movies['popularity'].median()
median_popularity
```

```python
#the most frequent genre of movies
genre_counts = movies['genres'].value_counts()
genre_counts
```

```python
#the average vote average of all movies
average_vote_average = np.mean(movies['vote_average'])
average_vote_average
```

```python
#the release year of the oldest movie
oldest_movie_year = movies['release_date'].min()
oldest_movie_year.date()
```

```python
#numbers of movies have a vote average greater than 8
high_vote_average_count = movies[movies['vote_average'] > 8]['vote_average'].count()
high_vote_average_count
```

```python
#the title of the movie with the longest runtime
longest_runtime_movie = movies.loc[movies['runtime'].idxmax(), 'title']
longest_runtime_movie
```

```python
#the title of the movie with the smallest runtime
smallest_runtime_movie = movies.loc[movies['runtime'].idxmin(), 'title']
smallest_runtime_movie
```

```python
# the average runtime of movies grouped by genre
average_runtime_by_genre = movies.groupby('genres')['runtime'].mean().sort_values(ascending=False)
average_runtime_by_genre
```

```python
#the top 10 highest-grossing movies (revenue - budget)
movies['profit'] = movies['revenue'] - movies['budget']
top_5_highest_grossing = movies[['title', 'profit']].sort_values('profit', ascending=False).head(10)
top_5_highest_grossing
```

```python
#genre has the highest sum popularity score
genre_popularity = movies.groupby('genres')['popularity'].sum().sort_values(ascending=False).head(5)
genre_popularity
```

```python
#genre has the highest sum vote count
genre_vote_count = movies.groupby('genres')['vote_count'].sum().sort_values(ascending=False)
genre_vote_count
```

```python
#language has the highest popularity score for movies
language_popularity = movies.groupby('original_language')['popularity'].sum().sort_values(ascending=False)
language_popularity
```

```python
#the total revenue generated by movies in each genre
total_revenue_by_genre = movies.groupby('genres')['revenue'].sum().sort_values(ascending=False)
total_revenue_by_genre
```

```python
#top 10 movies with the highest budget
top_10_budget_movies = movies.sort_values('budget', ascending=False).head(10)[['title', "budget"]]
top_10_budget_movies
```

```python
#the average budget for movies released in each year
avg_budget_by_year = movies.groupby('release_year')[['budget']].mean().sort_values('budget', ascending=False)
avg_budget_by_year
```

```python
#the movies with a vote average greater than 8 and a budget over $100 million
movies_high_rated_high_budget = movies[(movies['vote_average'] > 8) & (movies['budget'] > 100000000)][['title' , 'revenue']]
movies_high_rated_high_budget
```

```python
#the median budget for movies in each original language
sum_budget_by_language = movies.groupby('original_language')[['budget' , "revenue"]].sum().sort_values("budget",ascending=False).head(10)
sum_budget_by_language
```

```python
movies.to_csv('movies_updated.csv', index=False)
```
